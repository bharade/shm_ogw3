```python


# Define file path
file_path = "input/reg-norm0-200/regression_norm_0-200.txt"

# Initialize lists
losses, maes, val_losses, val_maes = [], [], [], []

# Regular expression to extract values
pattern = re.compile(r"loss: ([\d\.]+) - mae: ([\d\.]+) - val_loss: ([\d\.]+) - val_mae: ([\d\.]+)")

# Read file and extract data
with open(file_path, "r") as file:
    for line in file:
        match = pattern.search(line)
        if match:
            loss, mae, val_loss, val_mae = map(float, match.groups())
            losses.append(loss)
            maes.append(mae)
            val_losses.append(val_loss)
            val_maes.append(val_mae)
```

```python
len(losses),len(maes),len(val_losses),len(val_maes)
```

```python
# Example: Plot the losses
import matplotlib.pyplot as plt
```

```python
# Print results
print(f"Losses: {losses[:5]} ...")  # Print first 5 values as a preview
print(f"MAEs: {maes[:5]} ...")
print(f"Validation Losses: {val_losses[:5]} ...")
print(f"Validation MAEs: {val_maes[:5]} ...")


epochs = range(1, len(losses) + 1)
plt.figure(figsize=(10, 5))
plt.plot(epochs[1:], losses[1:], label="Loss")
plt.plot(epochs, val_losses, label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.title("Training and Validation Loss")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(epochs[1:], maes[1:], label="MAE")
plt.plot(epochs, val_maes, label="Validation MAE")
plt.xlabel("Epochs")
plt.ylabel("Mean Absolute Error")
plt.legend()
plt.title("Training vs Validation MAE")
plt.show()
```
