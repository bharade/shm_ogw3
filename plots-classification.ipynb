```python


# Define file path
file_path = "input/classification-metrics/classification_0-200.txt"

# Initialize lists
losses, accuracy, val_losses, val_accuracy = [], [], [], []



# Regular expression to extract values
pattern = re.compile(r"accuracy: ([\d\.]+) - loss: ([\d\.]+) - val_accuracy: ([\d\.]+) - val_loss: ([\d\.]+)")

# Read file and extract data
with open(file_path, "r") as file:
    for line in file:
        match = pattern.search(line)
        if match:
            acc, loss, val_acc, val_loss = map(float, match.groups())
            losses.append(loss)
            accuracy.append(acc)
            val_losses.append(val_loss)
            val_accuracy.append(val_acc)
```

```python
len(losses),len(accuracy),len(val_losses),len(val_accuracy)
```

```python


# Define file path
file_path = "input/classification-metrics/classification_201-500.txt"

# Regular expression to extract values
pattern = re.compile(r"accuracy: ([\d\.]+) - loss: ([\d\.]+) - val_accuracy: ([\d\.]+) - val_loss: ([\d\.]+)")

# Read file and extract data
with open(file_path, "r") as file:
    for line in file:
        match = pattern.search(line)
        if match:
            acc, loss, val_acc, val_loss = map(float, match.groups())
            losses.append(loss)
            accuracy.append(acc)
            val_losses.append(val_loss)
            val_accuracy.append(val_acc)
```

```python
len(losses),len(accuracy),len(val_losses),len(val_accuracy)
```

```python
# Example: Plot the losses
import matplotlib.pyplot as plt
```

```python
# Print results
print(f"Losses: {losses[:5]} ...")  # Print first 5 values as a preview
print(f"Accuracy: {accuracy[:5]} ...")
print(f"Validation Losses: {val_losses[:5]} ...")
print(f"Validation Accuracy: {val_accuracy[:5]} ...")


epochs = range(1, len(losses) + 1)
plt.figure(figsize=(10, 5))
plt.plot(epochs[1:], losses[1:], label="Loss")
plt.plot(epochs, val_losses, label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.title("Training and Validation Loss")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(epochs, accuracy, label="Training Accuracy")
plt.plot(epochs, val_accuracy, label="Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.title("Training vs Validation Accuracy")
plt.show()
```
