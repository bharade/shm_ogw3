```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_signal(csv_path, column_index, num_timesteps):
    """
    Plots the specified column of a CSV file against timesteps.

    Parameters:
        csv_path (str): Path to the CSV file.
        column_index (int): The column index (1-based) to plot.
        num_timesteps (int): Number of timesteps to consider.
    """
    # Load the CSV file
    data = pd.read_csv(csv_path)
    
    # Convert 1-based index to 0-based
    column_data = data.iloc[:num_timesteps, column_index - 1]

    # Generate timesteps
    timesteps = range(len(column_data))

    # Plot the data
    plt.figure(figsize=(10, 5))
    plt.plot(timesteps, column_data, label=f"Sensor Response")
    plt.xlabel("Timesteps")
    plt.ylabel("Sensor Response")
    plt.title(f"Sensor Response vs. Timesteps")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Example usage
csv_file = "signal-plot-samples/data_all_20191107T084901_baseline_1_40kHz.csv"
plot_signal(csv_file, column_index=27, num_timesteps=13108)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_signal_dam(csv_path, column_index, num_timesteps):
    """
    Plots the specified column of a CSV file against timesteps.

    Parameters:
        csv_path (str): Path to the CSV file.
        column_index (int): The column index (1-based) to plot.
        num_timesteps (int): Number of timesteps to consider.
    """
    # Load the CSV file
    data = pd.read_csv(csv_path)
    
    # Convert 1-based index to 0-based
    column_data = data.iloc[:num_timesteps, column_index - 1]

    # Generate timesteps
    timesteps = range(len(column_data))

    # Plot the data
    plt.figure(figsize=(10, 5))
    plt.plot(timesteps, column_data, color='red', label=f"$T_3-T_9$ Sensor Response")
    plt.xlabel("Timesteps")
    plt.ylabel("Sensor Response")
    plt.title(f"Damage Signal")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Example usage
csv_file = "signal-plot-samples/data_all_20191114T142828_D1_S9_R1_40kHz.csv"
plot_signal_dam(csv_file, column_index=27, num_timesteps=13108)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_signals(csv_path1, csv_path2, column_index, num_timesteps):
    """
    Plots the specified column from two CSV files against timesteps on the same graph.

    Parameters:
        csv_path1 (str): Path to the first CSV file.
        csv_path2 (str): Path to the second CSV file.
        column_index (int): The column index (1-based) to plot.
        num_timesteps (int): Number of timesteps to consider.
    """
    # Load the CSV files
    data1 = pd.read_csv(csv_path1)
    data2 = pd.read_csv(csv_path2)

    # Convert 1-based index to 0-based
    signal1 = data1.iloc[:num_timesteps, column_index - 1]
    signal2 = data2.iloc[:num_timesteps, column_index - 1]

    # Generate timesteps
    timesteps = range(num_timesteps)

    # Plot the data
    plt.figure(figsize=(10, 5))
    plt.plot(timesteps, signal1, color='#2F6DB3', label=f"Baseline Sample", alpha=0.7)
    plt.plot(timesteps, signal2, color='red', label=f"Damage Sample", alpha=0.7)
    plt.xlabel("Timesteps")
    plt.ylabel("Sensor Response")
    plt.title(f"Sensor Response vs. Timesteps")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python

# Example usage
csv_file1 = "signal-plot-samples/data_all_20191107T084901_baseline_1_40kHz.csv"
csv_file2 = "signal-plot-samples/data_all_20191114T142828_D1_S9_R1_40kHz.csv"

plot_signals(csv_file1, csv_file2, column_index=27, num_timesteps=13107)
```
